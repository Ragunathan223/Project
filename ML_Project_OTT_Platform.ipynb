### Data Preprocessing

```python
df.rename(columns={
    'How many OTT platforms do you use?': 'No. of platforms using',
    'Which OTT Platform do you prefer?': 'Preferred platform',
    'Which is your preferred language to watch TV shows on the OTT \nPlatforms?': 'Preferred language',
    'Do you think that the OTT platform\'s recommendations have influenced your viewing habits?': 'Viewing habits influence',
    'On average, how much time do you spend on watching content through OTT \nplatforms in day?': 'Avg. time spent',
    'How often do you use OTT platforms to stream movies or TV shows?': 'Often usage',
    ' What kind of Entertainment would you prefer the most to watch on the OTT \nplatforms?': 'Preference',
    'Which of these devices do you use to consume content through OTT platforms?': 'Devices used',
    'Have you ever cancelled a subscription to an OTT platform due to the high cost?': 'Cancelled subscription due to high cost',
    'Have you ever shared your OTT platform account with friends or family members?': 'Shared Account info',
    'Have you ever been dissatisfied with the availability of content on your preferred OTT platform(s)?': 'Dissatisfied with availability of content',
    'Which plan do you prefer?': 'Plan preference',
    'Have you ever subscribed to an OTT platform just to watch a specific show or movie?': 'Subscribed to watch specific content',
    'Have you ever cancelled a subscription to an OTT platform due to technical difficulties or poor video streaming quality?': 'Cancelled subscription due to streaming quality',
    'What factors do you consider when choosing an OTT platform to subscribe to?': 'Factors considered',
    'Have you ever signed up for a free trial on the OTT platform?': 'Free trial',
    'Do you think the pricing for the OTT platform is reasonable?': 'Reasonable price',
    'Have you ever switched from one OTT platform to another due to a better selection of content or features?': 'Switched platform',
    'Have you ever used an OTT platform to watch content that is not available in your country or region?': 'Content not available in your country',
    'Have you ever used a VPN to access content not available in your region on the OTT platform?': 'VPN access for content unavailable',
    'Have you ever used the ratings and reviews feature on the OTT platform to make a decision on what to watch?': 'Rating & reviews for decision',
    'Have you ever experienced buffering or connectivity issues while streaming on the OTT platform?': 'Buffering while streaming',
    'How satisfied are you with the availability of subtitles and closed captions on the OTT platform(s) you use?': 'Satisfied with availability of subtitles & closed captions',
    'Did you use any OTT platforms before covid?': 'Before Covid',
    'Do you think the habit of watching OTT based video content has increased during \nthe lock-down period?': 'Increase usage during lock down',
    'Do you prefer watching movies on OTT platforms instead of watching movies in a \ncinema hall?': 'OTT instead of Cinema Hall',
    'Do you think watching movies in OTT platforms are cheaper and more convenient \nthan other traditional media?': 'Cheaper than traditional media',
    'Would you go to the cinema hall to watch a movie with your friends and family as \nfrequently as before?': 'Frequent theatre visit as before'
                  }, inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df['Cancelled subscription due to high cost'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Shared Account info'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Dissatisfied with availability of content'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Subscribed to watch specific content'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Cancelled subscription due to streaming quality'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Free trial'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Reasonable price'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Switched platform'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Content not available in your country'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['VPN access for content unavailable'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Rating & reviews for decision'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Buffering while streaming'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Before Covid'].replace(('No', 'Yes'), (0, 1), inplace=True)
df['Cheaper than traditional media'].replace(('No', 'Yes'), (0, 1), inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df['Satisfied with availability of subtitles & closed captions']=df['Satisfied with availability of subtitles & closed captions'].fillna(df['Satisfied with availability of subtitles & closed captions'].mode())
```
